# TensorBoard的使用
* 直接使用Neural Network中用到的网络结构
* 可以对调参结果进行对比

```python
# 导入MNIST数据
from tensorflow.examples.tutorials.mnist import input_data
mnist = input_data.read_data_sets("/tmp/data/", one_hot=True)
# Imoprt TensorFlow
import tensorflow as tf
```

```python
# 训练用参数
learning_rate = 0.01
num_steps = 500
batch_size = 100
display_step = 100
initializer = tf.random_uniform_initializer()
# 神经网络参数
n_hidden_1 = 256
n_hidden_2 = 128
num_input = 28*28  #图像的大小
num_classes = 10  #0~9共10个数字
# 输入的placeholder
X = tf.placeholder(tf.float32, [None, num_input])
Y = tf.placeholder(tf.float32, [None, num_classes])
with tf.variable_scope("weights_and_biases"):
    # 矩阵W
    weights_h1 = tf.get_variable(name = "W1", shape = [num_input, n_hidden_1], initializer = initializer)
    weights_h2 = tf.get_variable(name = "W2", shape = [n_hidden_1, n_hidden_2], initializer = initializer)
    weights_out = tf.get_variable(name = "W_out", shape = [n_hidden_2, num_classes], initializer = initializer)
    # 向量b
    bias_h1 = tf.get_variable(name = "b1", shape = [n_hidden_1], initializer = initializer)
    bias_h2 = tf.get_variable(name = "b2", shape = [n_hidden_2], initializer = initializer)
    bias_out = tf.get_variable(name = "b_out", shape = [num_classes], initializer = initializer)
# f(XW+b)
# 构建神经网络
with tf.variable_scope("nerual_network"):
    layer_1 = tf.nn.relu(tf.nn.xw_plus_b(X, weights_h1, bias_h1))
    layer_2 = tf.nn.relu(tf.nn.xw_plus_b(layer_1, weights_h2, bias_h2))
    logits = tf.nn.relu(tf.nn.xw_plus_b(layer_2, weights_out, bias_out))
# 建立loss function
loss = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(
                                logits = logits, labels = Y))
optimizer = tf.train.AdamOptimizer(learning_rate = learning_rate)
train_op = optimizer.minimize(loss)
# 评估用model
correct_pred = tf.equal(tf.argmax(logits, 1), tf.argmax(Y, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
# 记录保存位置
log_dir = 'D:/tfNote/TensorBoard/logs'

# 创建记录
train_loss_summary = tf.summary.scalar('train_cost',loss)
train_accuracy_summary = tf.summary.scalar('train_accuracy', accuracy)
test_loss_summary = tf.summary.scalar('test_loss',loss)
test_accuracy_summary = tf.summary.scalar('test_accuracy', accuracy)

summary_op_train=tf.summary.merge([train_loss_summary, train_accuracy_summary])
summary_op_test=tf.summary.merge([test_loss_summary, test_accuracy_summary])
```

* 也可以使用tf.summary.merge_all() 这时不需要将tf.summary.scalar()创建的记录赋值给某个变量

```python
# 初始化
init = tf.global_variables_initializer()
# 训练网络
with tf.Session() as sess:
    sess.run(init)
    
    summary_writer = tf.summary.FileWriter(log_dir,sess.graph) # 写下记录用
    
    for step in range(num_steps):
        batch_x, batch_y = mnist.train.next_batch(batch_size)
        
        _, train_summary_in_step = sess.run([train_op, summary_op_train], feed_dict = {X:batch_x, Y:batch_y})
        summary_writer.add_summary(train_summary_in_step, step)
        
        if step%display_step == 0 :
            Loss, acc = sess.run([loss, accuracy], feed_dict={X:batch_x,Y:batch_y})
            print("Step " + str(step) + ", Minibatch Loss= " + \
                   "{:.4f}".format(Loss) + ", Training Accuracy= " + \
                   "{:.3f}".format(acc))
            
            test_accuracy, test_summary = sess.run([accuracy, summary_op_test], feed_dict={X: mnist.test.images, Y: mnist.test.labels})
            summary_writer.add_summary(test_summary, step)
            
            print("Testing Accuracy:",test_accuracy)
            
        print("Optimization Finished!")
        

print("All Finished!")
```
